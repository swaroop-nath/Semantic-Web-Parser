It seems that __Num Formatting Tags__ and __Num Words__ add appreciable distinction in the border cases of __Fraction p Children__. It can be used in the future, if there is a necessity.

Let's now fit a Logistic Regression model to the data and see the outcome. However, we will verify if the assumption - "__The log odds of success varies linearly with the independent variables__".
The above assumption states that the probability of getting on of the class will linearly increase with the independent variables.

```python
sns.scatterplot(x = 'Fraction p Children', y = 'Label', hue = 'Label', data = dataset)
plt.show()
sns.scatterplot(x = 'Num Words', y = 'Label', hue = 'Label', data = dataset)
plt.show()
```

__Fraction p Children__ does abide by the assumption for Logistic Regression. However, __Num Words__ doesn't seem to do so. Let's see if a convolution of __Num Words__ and __Fraction p Children__ seems to abide by it.

```python
dataset['Add'] = dataset.apply(lambda row: row['Fraction p Children'] + row['Num Words'], axis = 1)
dataset['Multiply'] = dataset.apply(lambda row: row['Fraction p Children'] + row['Num Words'], axis = 1)
```

```python
sns.pairplot(dataset, vars = ['Fraction p Children', 'Add', 'Multiply'], hue = 'Label')
plt.show()
sns.scatterplot(x = 'Add', y = 'Label', hue = 'Label', data = dataset)
plt.show()
sns.scatterplot(x = 'Multiply', y = 'Label', hue = 'Label', data = dataset)
plt.show()
```

Both multiplication and addition do tend to create a better variable than __Num Words__ itself (abiding better by the assumption of Logistic Regression), keeping the discrimation power similar. I will be using the multiplicative interaction for further analysis.

```python
def sigmoid(X, theta_1, theta_2, a, b):
    X_1, X_2 = X
    return 1/(1 + np.exp(-a * (theta_1 * X_1 + theta_2 * X_2 - b)))
```

```python
X['Fraction p Children']
```

```python
X_1 = dataset['Fraction p Children'].values
X_2 = dataset['Multiply'].values
y = dataset['Label'].values

params_optimal, _ = optimize.curve_fit(sigmoid, xdata = (X_1, X_2), ydata = y, p0 = (1, 1, 1, 0.1))

pseudo_X = np.arange(0, 1, 0.05) 
y_proba = sigmoid(pseudo_X, *params_optimal)

dataset['Combination'] = dataset.apply(lambda row: params_optimal[0] * row['Fraction p Children'] 
                                       + params_optimal[1] * row['Multiply'], axis = 1)

sns.scatterplot(x = 'Combination', y = 'Label', hue = 'Label', data = dataset)
plt.plot(pseudo_X, y_proba, label = 'Fitted Data')
plt.legend()
```

```python
X = dataset['Fraction p Children'].values
y = dataset['Label'].values

sc = StandardScaler()
X = sc.fit_transform(X.reshape(-1, 1))

model = LogisticRegression().fit(X, y)

pseudo_X = np.arange(min(X), max(X), 0.05)
y_proba = model.predict_proba(pseudo_X.reshape(-1, 1))

sns.scatterplot(x = list(X.reshape(1, -1)[0]), y = y)
plt.plot(pseudo_X, y_proba, label = 'Fitted Data')
plt.hlines(y = 0.5, xmin = -1, xmax = 20, linestyle = '--', color = 'red')
plt.legend()
```
